```python
import matplotlib.pyplot as plt

data = [
    {"job_title": "Senior Machine Learning Engineer", "salary_year_avg": 325000.0},
    {"job_title": "Principal Machine Learning Engineer", "salary_year_avg": 325000.0},
    {"job_title": "Machine Learning Engineer", "salary_year_avg": 300000.0},
    {"job_title": "Principal ML Research Scientist", "salary_year_avg": 262500.0},
    {"job_title": "Machine Learning Researcher", "salary_year_avg": 262500.0},
    {"job_title": "Director of Engineering, ML Platform", "salary_year_avg": 249000.0},
    {"job_title": "Machine Learning Engineer", "salary_year_avg": 240000.0},
    {"job_title": "Sr. Machine Learning Engineer", "salary_year_avg": 225000.0},
    {"job_title": "Director of Engineering, ML Platform", "salary_year_avg": 222000.0},
    {"job_title": "Lead AI / ML / Data Science Engineer", "salary_year_avg": 200000.0}
]

job_titles = [job["job_title"] for job in data]
salaries = [job["salary_year_avg"] for job in data]

plt.figure(figsize=(10, 6))
plt.barh(job_titles, salaries, color='skyblue')
plt.xlabel('Average Salary ($)')
plt.title('Average Salary distribution for Top Ten Machine Learning jobs in 2023')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
